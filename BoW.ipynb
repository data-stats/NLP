```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

#Consider the 3 documents containing below content
doc1 = "Natural Language processing heps us in playing with words in any language."
doc2 = "Natural language processing is ne important concept in machine learning use cases"
doc3 = "We are here to help people to upgrade their skills and become expert in demanding technologies"

vectr = CountVectorizer()
#CountVectorizer is used to tokenize a collection of text documents and build a known vocabulary of known words. It returns integer count of each word found in document.
corpus = [doc1,doc2,doc3]
cntvect = vectr.fit_transform(corpus)
print(cntvect)
```

```python
df = pd.DataFrame(cntvect.toarray().transpose(),index = vectr.get_feature_names())
df.columns = ['document1','document2','document3']
df
```

```python
vectr = TfidfVectorizer() # TfidfVectorizer is a class that returns TF-IDF value of each word in each document.
cntvect = vectr.fit_transform(corpus)
print(cntvect)
```

```python
df = pd.DataFrame(cntvect.toarray().transpose(),index = vectr.get_feature_names())
df.columns = ['document1','document2','document3']
df
```
